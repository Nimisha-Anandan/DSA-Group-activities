### ANOVA test

Testing whether the same anount was spemt for three adds

**H0:Mean amount spent for all the three adds are equal**

**H1:Amount spent for atleast one add differes significantly**

```python
from scipy.stats import f_oneway
f_oneway(df['TV'],df['Radio'],df['Newspaper'])
```

Here p value is 0. Since p value < 0.05 we reject the null hypothesis

**We can infer that amount spent on the three adds are different**
